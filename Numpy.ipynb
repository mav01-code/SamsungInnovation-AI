```python
arr1 = np.array([1, 3, 5, 7, 9])
print(arr1)
arr2 = np.array([1, 3, 5, 7, 9])
print(arr2)
print(type(arr1))
print(id(arr1))
```

```python
# Shallow copy - both arr3 and arr1 point to the same object
arr3 = arr1
print(id(arr3))
print(id(arr1))
```

```python
# Deep copy - arr4 is a new object
arr4 = arr1.copy()
print(id(arr4))
print(id(arr1))
```

```python
a = np.arange(10) # Shallow copy
print(a)
b = a.reshape(2, 5)
print(b)
a[0] = 999
print(a)
print(b)
```

```python
c = a.reshape(2, 5).copy() # Deep copy
c[0, 0] = 0
print(a)
print(c)
```
