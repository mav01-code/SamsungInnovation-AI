```python
# Tokenize into sentences
sentences = nltk.sent_tokenize(paragraph)
print(sentences)
```

### 1.1 Tokenization

```python
# Tokenize into words
words = nltk.word_tokenize(paragraph)
print(words)
```

### 1.2 Stemming
- Stemming reduces a word to its root form by rule-based stripping, not by dictionary meaning.
- Resulting stems may not be valid English words
- Eg: running - run

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
```

- SnowballStemmer - Language-aware stemmer
- LancasterStemmer - Very aggressive, shortens words heavily
- PorterStemmer - Most commonly used, moderate aggressiveness, balanced

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(x) for x in words]
    sentences[i] = ' '.join(words)
```

```python
print(sentences)
```

### 1.3 Lemmatization
- It reduces a word to its base dictionary from (lemma) using vocabulary and grammar.
- The output is always a valid English word

- Requires part-of-speech (POS) information
- Applies morphological analysis

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(x) for x in words]
  sentences[i] = ' '.join(words)
print(sentences)
```

```python
lemmatizer.lemmatize("wolves", wordnet.NOUN)
```

```python
lemmatizer.lemmatize("cried", wordnet.VERB)
```

```python
lemmatizer.lemmatize("happily", wordnet.ADJ)
```

- Stemming - Rule-based suffix stripping, May be non-words - Use for Large-scale IR systems, When speed matters more than linguistics
- Lemma - Dictionary + grammar-based - If semantics matter

```python
import re
```

```python
sentences = nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    sentences[i] = sentences[i].lower()
    sentences[i] = re.sub(r'\W', ' ', sentences[i])
    sentences[i] = re.sub(r'\s', ' ', sentences[i])
    sentences[i] = re.sub(r'\s+', ' ', sentences[i])
    sentences[i] = re.sub(r'\s$', ' ', sentences[i])
print(sentences)
```

### 2.1 Stop word removal
- NLTK provides predefined stop words

```python
from nltk.corpus import stopwords
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [x for x in words if x not in stopwords.words('english')]
  sentences[i] = ' '.join(words)
print(sentences)
```

### POS Tagging
- It assigns a grammatical label to each word in a sentence
Common tags: 
1) NN Noun
2) VB verb
3) JJ adjective
4) RB adverb

- If we don't perform or do POS tagging, many sentences lose it's semantic accuracy.

```python
my_sentence = "The Colosseum was built by the emperor Vespassian"
my_sentence = "The Colosseum was built by the emperor Vespassian"
my_words = nltk.word_tokenize(my_sentence)
for i in range(len(my_words)):
  my_words[i] = my_words[i].lower()
print(my_words)
```
